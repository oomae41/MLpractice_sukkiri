```python
import  pandas as pd

df = pd.read_csv("../sukkiri-ml-codes/datafiles/ex4.csv")
```

```python
df
```

```python
# 男性比率の計算
df['sex'].mean()
```

```python
# 役職ごとのscoreの平均
df.groupby('class')['score'].mean()
```

```python
# 役職と性別ごとのscoreの平均をクロス集計
pd.pivot_table(df, index="class", columns="sex", values="score", aggfunc="mean")
```

```python
# dept_idのダミー変数化
dept = pd.get_dummies(df['dept_id'], drop_first = True, prefix='dept', dtype=int)
dept
```

```python
df2 = pd.read_csv("../sukkiri-ml-codes/datafiles/ex4.csv")

# df2にdeptを連結してdept_idを削除
df2 = pd.concat([df2, dept], axis=1).drop('dept_id', axis=1)
```
