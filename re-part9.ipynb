```python
import pandas as pd
df = pd.read_csv("..\サンプルデータ\sukkiri-ml-codes\datafiles\Bank.csv")
df.head()
```

```python
%matplotlib
from sklearn.model_selection import train_test_split
```

```python
# 文字列データを一気にダミー化
str_col = ["job", "marital", "education", "default", "housing", "loan", "contact", "month"]
str_col_df = pd.get_dummies(df[str_col], drop_first=True, dtype=int)
df2 = pd.concat([df, str_col_df], axis=1)
df2.head()
```

```python
df2.columns
```

```python
#訓練&検証データとテストデータに分割
train_val,test = train_test_split(df2,test_size=0.2,random_state=0)
train_val.head()
```

```python
# durationの欠損値を平均で補完
duration_mean = train_val["duration"].mean()
train_val2 = train_val.fillna(duration_mean)
train_val2.isnull().sum()
```

```python
# 文字列データを除外
x = train_val2.drop(str_col, axis=1)
# 特徴量としていらない列も除外
x = x.drop(['id','y','day'],axis=1)
# 目的変数
t = train_val2["y"]

# train_val2を訓練用と検証用に分割
x_train, x_test, y_train, y_test = train_test_split(x, t, test_size=0.2, random_state=0)
```

```python
# 決定木モデルの作成
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth=5, random_state=0, class_weight="balanced")
model.fit(x_train, y_train)
```

```python
# 同じ処理を繰り返すのでlearn関数を定義
def learn(x,t,depth):
    x_train,x_val,y_train,y_val = train_test_split(x,t,test_size=0.2,random_state=13)

    datas=[x_train,x_val,y_train,y_val]

    model = tree.DecisionTreeClassifier(random_state=depth,max_depth=depth,class_weight='balanced')
    model.fit(x_train,y_train)
    train_score=model.score(x_train,y_train)
    
    val_score=model.score(x_val,y_val)
    return train_score,val_score,model,datas
```

```python
# 深さを変えてモデル作成
for i in range(1,20):
    s1,s2,model,datas = learn(x,t,i)
    print(i,s1,s2)

# サンプルと同じ結果を得られる
```

```python
# 深さ10が最大性能だった

#特徴量重要度の確認
pd.Series(model.feature_importances_,index=x.columns).sort_values(ascending=False)

# やはりdurationが強いことがわかる
```

```python
# 穴埋め処理を関数として定義
def nan_fill(train_val):
    isnull = train_val['duration'].isnull()

    train_val2 = train_val.copy()
    #housing=yesの処理
    train_val2.loc[(isnull) & (train_val2['housing']=='yes') 
                  &(train_val2['loan']=='yes'),'duration' ]=439
    train_val2.loc[(isnull) & (train_val2['housing']=='yes') 
                  &(train_val2['loan']=='no'),'duration' ]=332

    #housing=noの処理
    train_val2.loc[(isnull) & (train_val2['housing']=='no') 
                  &(train_val2['loan']=='yes'),'duration' ]=301
    train_val2.loc[(isnull) & (train_val2['housing']=='no') 
                  &(train_val2['loan']=='no'),'duration' ]=237
    
    return train_val2
```

```python
# train_val2を変更
train_val2=nan_fill(train_val)
```

```python
t =train_val2['y']
x = train_val2.drop(str_col,axis=1)
x =x.drop(['id','y','day'],axis=1)

for i in range(1,20):
    s1,s2,model,datas = learn(x,t,i)
    print(i,s1,s2)
```

```python
# テストデータでスコア確認
model_tree=tree.DecisionTreeClassifier(max_depth=10,random_state=10,class_weight="balanced")
model_tree.fit(x,t)

test2=nan_fill(test)
t =test2['y']
x = test2.drop(str_col,axis=1)
x =x.drop(['id','y','day'],axis=1)
model_tree.score(x,t)
```
