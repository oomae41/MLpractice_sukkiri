ダミー変数化（その後データを分割）

```python
# jobをダミー変数化
job_dummy = pd.get_dummies(df["job"], drop_first=True, dtype=int)

# dfとjob_dummyを連結
df2 = pd.concat([df, job_dummy], axis=1)

# df2から元のjobを削除
df2 = df2.drop(["job"], axis=1)

df2.head()
```

```python
# maritalをダミー変数化
marital_dummy = pd.get_dummies(df["marital"], drop_first=True, dtype=int)

# marital_dummyを連結
df3 = pd.concat([df2, marital_dummy], axis=1)

# df2から元のmaritalを削除
df3 = df3.drop(["marital"], axis=1)

df3.head()
```

```python
# educationをダミー変数化
education_dummy = pd.get_dummies(df["education"], drop_first=True, dtype=int)

# marital_dummyを連結
df4 = pd.concat([df3, education_dummy], axis=1)

# 元のeducationを削除
df4 = df4.drop(["education"], axis=1)

df4.head()
```

```python
# contactをダミー変数化
education_dummy = pd.get_dummies(df["contact"], drop_first=True, dtype=int)

# contact_dummyを連結
df5 = pd.concat([df4, education_dummy], axis=1)

# 元のcontactを削除
df5 = df5.drop(["contact"], axis=1)

df5.head()

# drop_first=Trueなので元の変数の種類より1つ少ない列数のDFが返ってくる
# すべて0であれば消した列が1になっているという読み方をする
```

```python
# defaultをダミー変数化
default_dummy = pd.get_dummies(df["default"], drop_first=True, dtype=int)

# default_dummyを連結
df6 = pd.concat([df5, default_dummy], axis=1)

# 元のdefaultを削除
df6 = df6.drop(["default"], axis=1)

df6.head()
```

データの分割

```python
# 訓練＋検証データとテストデータに分割
# testは最後のモデル評価のみに使う
# train_valは加工後に再び分割する
train_val, test = train_test_split(df5, test_size=0.2, random_state=0)
```

決定木モデルの作成（カテゴリカルデータも欠損値も扱える）

```python
# 特徴量の選択
# default, housing, loan, monthはダミー化してないから除外
# dayと2つのunknownも除外 
col = ['age', 'amount', 
       'duration', 'campaign', 'previous', 'blue-collar', 'entrepreneur',
       'housemaid', 'management', 'retired', 'self-employed', 'services',
       'student', 'technician', 'unemployed', 'married', 'single',
       'secondary', 'tertiary', 'sending _document', 'telephone']

x = train_val[col]
t = train_val["y"]
```

```python
# 分割
x_train, x_test, y_train, y_test = train_test_split(x, t, test_size=0.2, random_state=0)
```

```python
# 決定木モデルの作成
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth=5, random_state=0, class_weight="balanced")
model.fit(x_train, y_train)
```

```python
# 正解率を計算
model.score(x_test, y_test)
```

モデルの改善

```python
# 同じ処理を繰り返すのでlearn関数を定義
def learn(x,t,depth):
    x_train,x_val,y_train,y_val = train_test_split(x,t,test_size=0.2,random_state=13)

    datas=[x_train,x_val,y_train,y_val]

    model = tree.DecisionTreeClassifier(random_state=depth,max_depth=depth,class_weight='balanced')
    model.fit(x_train,y_train)
    train_score=model.score(x_train,y_train)
    
    val_score=model.score(x_val,y_val)
    return train_score,val_score,model,datas
```

```python
# 深さを変えてモデル作成
for i in range(1,20):
    s1,s2,model,datas = learn(x,t,i)
    print(i,s1,s2)
```

特徴量の取り方を解答に合わせる
